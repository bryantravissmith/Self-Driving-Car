```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

c0 = 0
c1 = 1

def poly(x):
    return c0 + c1 * x

x = np.linspace(-10,10)
y = poly(x)

ptx = 2
pty = -3

closest_x = (pty * c1 - c1 * c0 + ptx) / (1 + c1**2)
closest_y = poly(closest_x)

plt.figure(figsize=(8,8))
plt.plot(x,y)
plt.scatter(ptx, pty)
plt.scatter(closest_x, closest_y, c='red')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
